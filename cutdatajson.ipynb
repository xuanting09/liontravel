```python
import json

# 設定切割大小
lines_per_file = 10000  # 可自行調整

# 讀取 JSON 檔案
input_file = "log_202409.json"

with open(input_file, "r", encoding="utf-8") as f:
    data = json.load(f)  # 解析 JSON 檔案

# 計算總數據筆數
total_lines = len(data)
print(f"總行數: {total_lines}")

# 切割 JSON
for i in range(0, total_lines, lines_per_file):
    part_data = data[i:i + lines_per_file]  # 取部分數據
    part_file = f"log_202409_part{i//lines_per_file+1}.json"

    with open(part_file, "w", encoding="utf-8") as f:
        json.dump(part_data, f, ensure_ascii=False, indent=4)  # 寫入拆分檔案
    
    print(f"已生成: {part_file}")

print("完成 JSON 拆分！")
```

```python
import json
import pandas as pd

# 讀取 JSON 檔案
file_path = "log_202409_part1.json"

with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# 解析關鍵字段
processed_data = []
for entry in data:
    logtime = entry.get("logtime", "")
    ptype = entry.get("ptype", "")
    luid = entry.get("luid", "")
    
    prod_info = entry.get("prod_info", {})
    prod_type = prod_info.get("ProdType", "")
    prod_name = prod_info.get("ProdName", "")
    prod_no = prod_info.get("ProdNo", "")
    prod_price = prod_info.get("ProdPrice", None)
    is_foreign = prod_info.get("IsForeign", None)

    processed_data.append({
        "logtime": logtime,
        "ptype": ptype,
        "luid": luid,
        "prod_type": prod_type,
        "prod_name": prod_name,
        "prod_no": prod_no,
        "prod_price": prod_price,
        "is_foreign": is_foreign
    })

# 轉換為 DataFrame
df = pd.DataFrame(processed_data)


# 儲存成 CSV 方便查看
df.to_csv("processed_log_data.csv", index=False, encoding="utf-8-sig")

# 顯示前 5 行數據
print(df.head())
```
